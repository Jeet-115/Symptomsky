# ***Implementation***

```python
loader=PyPDFLoader("/content/diabetes.pdf")
pages=loader.load_and_split()
pages
```

```python
from langchain.prompts import PromptTemplate

template="""Answer the question from the context below. If you cant find the answer,reply IDK

Context={context}
Question={question}

 """
prompt=PromptTemplate.from_template(template=template)
prompt.format(context="Here is the context",question="Here is the task")
```

```python
from langchain.schema import Document
from langchain_community.vectorstores import DocArrayInMemorySearch
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")# embeddings = [embedding_model.encode(page.page_content) for page in pages]

vec_store=DocArrayInMemorySearch.from_documents(
    pages,
    embedding=embeddings
)

# Create a DocArrayInMemorySearch vector store
# vec_store = DocArrayInMemorySearch.from_documents(pages,embedding_model)
```

```python
retreiver=vec_store.as_retriever()
#retreiver.invoke("Iterated Hash Function")
```

```python
from operator import itemgetter
chain=(
    {"context":itemgetter('question') | retreiver ,
    "question":itemgetter('question')} | prompt | groq_chat | parser
)

# chain.invoke({"question":"What is an Iterated hash function?"})
```
